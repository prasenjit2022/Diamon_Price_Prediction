```python
import pandas as pd
```

```python
## Data Ingestion steps
df=pd.read_csv('cubic_zirconia.csv')
df.head()
```

```python
## Check any missing values in dataset.
df.isnull().sum()
```

## Observation: depth field is having 697 missing values

```python
df.info()
```

```python
## Drop the id column from the dataset
df=df.drop(labels=['id'],axis=1)
df.head()
```

```python
## Check for duplicate records
df.duplicated().sum()
```

## There are 34 records having duplicate values we need remove duplicate values


```python
## Dropping the duplicate records
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

## All the duplicate values are removed.

```python
## Seggregate numerical and categorical columns
numerical_columns=df.columns[df.dtypes!='O']
categorical_columns=df.columns[df.dtypes=='O']
print("Numerical columns:",numerical_columns)
print("Categorical columns:",categorical_columns)
```

```python
df[categorical_columns].describe()
```

```python
df['cut'].value_counts()
```

```python
df['clarity'].value_counts()
```

```python
df['color'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
x=0
for i in numerical_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```

```python
## Categorical features EDA
plt.figure(figsize=(8,6))
x=0
for i in categorical_columns:
    sns.catplot(data=df,x='cut',y=df.count)
    print('\n')
    plt.show()
```

```python
## Correlation
sns.heatmap(df[numerical_columns].corr(),annot=True)
```

```python
cut_map={"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}
```

```python
df['clarity'].unique()
```

```python
clarity_map={"I1":1,"SI2":2,"SI1":3,"VS2":4,"VS1":5,"VVS2":6,"VVS1":7,"IF":8}
```

```python
color_map={"D":1,"E":2,"F":3,"G":4,"H":5,"I":6,"J":7}
```

```python
df['cut']=df['cut'].map(cut_map)
df['clarity']=df['clarity'].map(clarity_map)
df['color']=df['color'].map(color_map)
```
